```python
import pandas as pd
import numpy as np


jugadores=pd.read_csv('CaracteristicasJ.csv')
tabla=pd.read_csv('TablaP.csv')
for i in range(len(tabla)):
    if tabla.loc[i]['Pos']=='P':
        tabla=tabla.drop(i)
for i in range(len(jugadores)):
    if jugadores.loc[i]['Pos']=='P':
        jugadores=jugadores.drop(i)

jugadores=jugadores.set_index(['Player','Pos'])
tabla=tabla.set_index(['Player','Pos'])
tabla
data=pd.merge(tabla,jugadores,how='inner',left_index=True,right_index=True)
data=data.reset_index()
for i in range(0,549):
    if data.loc[i, 'Edad'].count(':')>0:
        n = data.loc[i, 'Edad'].find(":")
        data.loc[i, 'Edad']=data.loc[i, 'Edad'][n+1:len(data.loc[i,'Edad'])]

    data['Throw']=data['Bateo']

for i in range(0,549):
    if data.loc[i, 'Bateo'].count(':')>0:
        n = data.loc[i, 'Bateo'].find(":")
        data.loc[i, 'Bateo']=data.loc[i, 'Bateo'][n+1:n+3]
    if data.loc[i,'Throw'].count('/')>0:
        data.loc[i,'Throw']=data.loc[i,'Throw'][7:8]
data=data.rename(columns={'Bateo':'Bat','Peso':'Weight','Edad':'Age'})
for i in range(0,549):
    if data.loc[i,'Bat']==' R':
        data.loc[i,'Bat']=1
    elif data.loc[i,'Bat']==' L':
        data.loc[i,'Bat']=-1
    elif data.loc[i,'Bat']==' S':
        data.loc[i,'Bat']=0
for i in range(0,549):
    if data.loc[i,'Throw']=='R':
        data.loc[i,'Throw']=1
    elif data.loc[i,'Throw']=='L':
        data.loc[i,'Throw']=-1
    elif data.loc[i,'Throw']=='S':
        data.loc[i,'Throw']=0
data['Height']=data['Weight']
for i in range(0,549):
    if data.loc[i, 'Weight'].count('/')>0:
        m = data.loc[i, 'Weight'].find("/")
        data.loc[i, 'Weight']=data.loc[i, 'Weight'][m+1:len(data.loc[i,'Weight'])]
for i in range(0,549):
    if data.loc[i, 'Height'].count('/')>0:
        q = data.loc[i, 'Height'].find("/")
        data.loc[i, 'Height']=data.loc[i, 'Height'][0:q]

for i in range(0,549):
    r=float(data.loc[i]['Height'][3])
    r=r/12
    h=(float(data.loc[i]['Height'][0])+r)/3.28
    t=round(h,2)
    data.loc[i,'Height']= t
for i in range(0,549):
    z=float(data.loc[i]['Weight'])
    m=z*0.453
    m=round(m)
    data.loc[i,'Weight']=m
data=data.drop(252)
I=list(range(0,548))
data=data.set_index([I])
for i in range(0,548):
    v=float(data.loc[i]['AVG'])
    b=float(data.loc[i]['SLG'])
    data.loc[i,'AVG']=v
    data.loc[i,'SLG']=b
    if data.loc[i]['OPS']>1:
        data.loc[i,'OPS']=data.loc[i]['OPS']/1000
    if data.loc[i]['Age']=='null':
        data.loc[i,'Age']=np.nan
    if data.loc[i]['GO_AO']=='-':
        data.loc[i,'GO_AO']=np.nan
data['Age']=data['Age'].fillna(26)
data['GO_AO']=data['GO_AO'].fillna(0)
data=data.drop(126)
J=list(range(0,547))
data=data.set_index([J])
data=data.set_index(['Player','Pos'])
data
data.to_csv('Datos.csv',sep=',',header=True,index=True)



    

    
    
    
```
